```python
df = pd.read_csv(r"C:\\Users\\Admin\\Desktop\\PW\\PredictivePacers\\Raw Data\\totalsixes.csv")
df.drop(['Unnamed: 4','4s','Match Date','Scorecard'],axis = 1, inplace =True)
df
```

```python
df_encoded = pd.get_dummies(df, columns=['Team 1','Team 2','Winner', 'Ground'])
df_encoded
```

```python
Y = df_encoded['6s']
Y
```

```python
X = df_encoded.drop(columns=['6s'])
X
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from xgboost import XGBRegressor
model = XGBRegressor()
```

```python
model.fit(x_train, y_train)
x_test
```

```python
y_pred = model.predict(x_test)
```
